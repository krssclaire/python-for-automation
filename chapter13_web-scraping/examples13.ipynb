```python
# get API key in .env (for privacy matter)
from dotenv import load_dotenv
import os

load_dotenv()
API_key = os.getenv("OPENWEATHER_API_KEY")
```

```python
# get city infos
import requests
from pathlib import Path

city_name = 'San Francisco'
state_code = 'CA'
country_code = 'US'

response = requests.get(f'https://api.openweathermap.org/geo/1.0/direct?q={city_name},{state_code},{country_code}&appid={API_key}')
response.text
```

```python
import json 
response_data = json.loads(response.text)
response_data
```

```python
# Find latitude and longitude
print(f"Latitude: {response_data[0]['lat']}")
print(f"Longitude: {response_data[0]['lon']}")
```

```python
# Find temperatures
lat = json.loads(response.text)[0]['lat']
lon = json.loads(response.text)[0]['lon']
response = requests.get(f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_key}')
```

```python
import json 
response_data = json.loads(response.text)
response_data

temp_k = response_data['main']['temp']
temp_c = round(temp_k -273.15, 1)
temp_f = round(temp_k * (9/5) - 459.67, 1)

print(f"Temperature in Kelvin (default): {temp_k}")
print(f"Temperature in Celsius: {temp_c}")
print(f"Temperature in Fahrenheit: {temp_f}")
```

```python
import requests, bs4

res = requests.get('https://autbor.com/example3.html')
res.raise_for_status()
example_soup = bs4.BeautifulSoup(res.text, 'html.parser')
elems = example_soup.select('#author')

print(type(example_soup))
print(example_soup.select('p #author'))

print(elems)
print(type(elems))
print(len(elems))
print(type(elems[0]))
print(str(elems[0]))
print(elems[0].get_text())
print(elems[0].attrs)
```

Selector passed to the select() method | Will match ...
---------------------------------------|---------------
soup.select('div')                     | All elements named <div>
soup.select('#author')                 | The element with an id attribute of author
soup.select('.notice')                 | All elements that use a CSS class attribute named notice
soup.select('div span')                | All elements named <span> that are within an element named <div>
soup.select('div > span')              | All elements named <span> that are directly within an element named <div>, with no other element in between
soup.select('input[name]')             | All elements named <\input> that have a name attribute with any value
soup.select('input[type="button"]')    | All elements named <\input> that have an attribute named type with the value button

```python
p_elems = example_soup.select('p')

print(f''' 
    {str(p_elems[0])}
    {p_elems[0].get_text()}
    {str(p_elems[1])}
    {p_elems[1].get_text()}
    {str(p_elems[2])}
    {p_elems[2].get_text()}
''')
```

### Image site downloader
Write a program that goes to a photo-sharing site like Flickr or Imgur, searches for a category of photo, and then downloads all the resulting images. You could write a program that works with any photo site that has a search feature.

```python
import requests, os
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# user inputs
category = 'fantasy'
MAX_DOWNLOADS = 5

# create folder in which images will be storec
os.makedirs(f'downloaded-{category.lower()}-imgs', exist_ok=True)

# find right category link
home_url = "https://books.toscrape.com/"
res = requests.get(home_url)
res.raise_for_status()
soup = BeautifulSoup(res.text, "html.parser")

category_links = soup.select(".nav-list ul li a")
category_url = None

for link in category_links:
    name = link.text.strip()

    if name.lower() == category.lower():
        category_url = urljoin(home_url, link["href"])
        break

print(f'Category url: {category_url}')

# opens page from category
res = requests.get(category_url)
res.raise_for_status()
soup = BeautifulSoup(res.text, "html.parser")

# find books
books = soup.select("article.product_pod")

# extract covers
for i, book in enumerate(books[:MAX_DOWNLOADS]):
    img = book.find("img")
    if img is None:
        continue
    img_url = urljoin(category_url, img["src"])
    img_res = requests.get(img_url)

    with open(f"downloaded-{category.lower()}-imgs/book{i+1}.jpg", "wb") as f:
        f.write(img_res.content)

print('Done, check directories')
```

### Link Verification
Write a program that, given the URL of a web page, will find every <\a> link on the page and test whether the linked URL results in a “404 Not Found” status code. The program should print out any broken links.


```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

print('Start checking...')
url = 'http://theodinproject.com/'
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
links = soup.find_all('a')

checked = set()
t = 0

for link in links:
    href = link.get('href')

    if not href:
        continue 
    if href.startswith("#") or href.startswith("mailto") or href.startswith("javascript"):
        continue

    full_url = urljoin(url, href)

    if full_url in checked:
        continue
    checked.add(full_url)

    r = requests.head(full_url)
    if r.status_code >= 400:
        print(f'Broken Link: {full_url}\nStatus code: {r.status_code}')
    
    t += 1

print(f'Done, checked {t} links')
```
