### Weather API


```python
# get API key in .env (for privacy matter)
from dotenv import load_dotenv
import os

load_dotenv()
API_key = os.getenv("OPENWEATHER_API_KEY")
```

```python
# get city infos
import requests
from pathlib import Path

city_name = 'San Francisco'
state_code = 'CA'
country_code = 'US'

response = requests.get(f'https://api.openweathermap.org/geo/1.0/direct?q={city_name},{state_code},{country_code}&appid={API_key}')
response.text
```

```python
import json 
response_data = json.loads(response.text)
response_data
```

```python
# Find latitude and longitude
print(f"Latitude: {response_data[0]['lat']}")
print(f"Longitude: {response_data[0]['lon']}")
```

```python
# Find temperatures
lat = json.loads(response.text)[0]['lat']
lon = json.loads(response.text)[0]['lon']
response = requests.get(f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_key}')
```

```python
import json 
response_data = json.loads(response.text)
response_data

temp_k = response_data['main']['temp']
temp_c = round(temp_k -273.15, 1)
temp_f = round(temp_k * (9/5) - 459.67, 1)

print(f"Temperature in Kelvin (default): {temp_k}")
print(f"Temperature in Celsius: {temp_c}")
print(f"Temperature in Fahrenheit: {temp_f}")
```

More about data in JSON format is at this link: [OpenWeater doc](https://openweathermap.org/api/one-call-3?collection=one_call_api_3.0&collection=one_call_api_3.0#current)

```python
# Get weather foreast
response = requests.get(f'https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API_key}')
response_data = json.loads(response.text)
response_data
```

### Parse HTML

```python
import requests, bs4

res = requests.get('https://autbor.com/example3.html')
res.raise_for_status()
example_soup = bs4.BeautifulSoup(res.text, 'html.parser')
elems = example_soup.select('#author')

print(type(example_soup))
print(example_soup.select('p #author'))

print(elems)
print(type(elems))
print(len(elems))
print(type(elems[0]))
print(str(elems[0]))
print(elems[0].get_text())
print(elems[0].attrs)
```

Selector passed to the select() method | Will match ...
---------------------------------------|---------------
soup.select('div')                     | All elements named <div>
soup.select('#author')                 | The element with an id attribute of author
soup.select('.notice')                 | All elements that use a CSS class attribute named notice
soup.select('div span')                | All elements named <span> that are within an element named <div>
soup.select('div > span')              | All elements named <span> that are directly within an element named <div>, with no other element in between
soup.select('input[name]')             | All elements named <\input> that have a name attribute with any value
soup.select('input[type="button"]')    | All elements named <\input> that have an attribute named type with the value button

```python
p_elems = example_soup.select('p')

print(f''' 
    {str(p_elems[0])}
    {p_elems[0].get_text()}
    {str(p_elems[1])}
    {p_elems[1].get_text()}
    {str(p_elems[2])}
    {p_elems[2].get_text()}
''')
```
