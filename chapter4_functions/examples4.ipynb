```python
def hello():
    # Prints three greetings
    print('Good morning!')
    print('Good afternoon!')
    print('Good evening!')

hello()
hello()
print('ONE MORE TIME!')
hello()
```

```python
def say_hello_to(name):
    # Prints three greetings to the name provided
    print('Good morning, ' + name)
    print('Good afternoon, ' + name)
    print('Good evening, ' + name)

say_hello_to('Alice')
say_hello_to('Bob')
say_hello_to('Al')
```

```python
def spam():
    eggs = 'spam local'
    print(eggs)  # Prints 'spam local'

def bacon():
    eggs = 'bacon local'
    print(eggs)  # Prints 'bacon local'
    spam()
    print(eggs)  # Prints 'bacon local'

eggs = 'global'
bacon()
print(eggs)  # Prints 'global'
```

```python
def spam():
    global eggs
    eggs = 'spam'  # This is the global variable.

def bacon():
    eggs = 'bacon'  # This is a local variable.

def ham():
    print(eggs)  # This is the global variable.

eggs = 'global'  # This is the global variable.
spam()
print(eggs)
```

```python
def spam(divide_by):
    try:
        # Any code in this block that causes ZeroDivisionError won't crash the program:
        return 42 / divide_by
    except ZeroDivisionError:
        # If ZeroDivisionError happened, the code in this block runs:
        print('Error: Invalid argument.')

print(spam(2))
print(spam(12))
print(spam(0))
print(spam(1))
```
