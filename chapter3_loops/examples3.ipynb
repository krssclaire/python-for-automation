# Chapter 3 - Loops

## Examples

```python
spam = 0
if spam < 5:
    print('Hello, world.')
    spam = spam + 1
```

```python
spam = 0
while spam < 5:
    print('Hello, world.')
    spam = spam + 1
```

```python
print('Hello!')
for i in range(5):
    print('On this iteration, i is set to ' + str(i))
print('Goodbye!')
```

```python
total = 0
for num in range(101):
    total = total + num
print(total)  
```

```python
print('Hello!')
i = 0
while i < 5:
    print('On this iteration, i is set to ' + str(i))
    i = i + 1
print('Goodbye!')
```

```python
for i in range(12, 16):
    print(i)
```

```python
for i in range(0, 10, 2):
    print(i)
```

```python
for i in range(5, -1, -1):
    print(i)
```

```python
s = 1
while s <= 10:
    print(s)
    s = s + 1
```

```python
import random
for i in range(5):
    print(random.randint(1, 10))
```
